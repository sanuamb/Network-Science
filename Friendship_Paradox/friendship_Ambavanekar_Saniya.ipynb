```python
#Measuring the friendship paradox
import networkx as nx

#Create a graph
first_graph=nx.Graph()

#adding nodes to the graph
first_graph.add_node(1)
first_graph.add_node(2)
first_graph.add_node(3)

#adding edges to the graph
first_graph.add_edge(1,2)
first_graph.add_edge(3,2)
```

```python
#Displaying the graph created
import matplotlib.pyplot as mp

mp.figure()

#Call draw method networkx to draw the graph
nx.draw(first_graph,with_labels=True)

#Display the graph
mp.show()
```

```python
#Finding how many nodes the graph has
num_nodes=len(first_graph)
print('The number of nodes :',num_nodes)

#Degree of any node in the graph
node3_degree=first_graph.degree(3)
print('The degree of node 3:',node3_degree)

#Finding the neighbours of any node
node2_neighbours=first_graph.neighbors(2)
print("The neighbours are:",[n for n in node2_neighbours])
```

```python
#Getting neighbours of all nodes in the sample graph
for node in first_graph.nodes():
    print("The neighbours are:",node,[n for n in first_graph.neighbors(node)])
```

```python
#Getting the degree of neighbor nodes for sample graph
import numpy as np

def find_avg(node,graph):
    n_neighbors=graph.neighbors(node)
    return (np.mean([graph.degree(n) for n in n_neighbors]))
    

for node in first_graph.nodes():
    deg=find_avg(node,first_graph)
    print("The avg_degree of neighbors for node",node,":",deg)
```

```python
#getting the friendship paradox for the sample graph

#if the avg_deg of neighbors greater than the respective degree of the node 
lst=[]
for node in first_graph:
    ideg=first_graph.degree(node)
    andeg=find_avg(node,first_graph)
    if (andeg>ideg):
        lst.append(1)
frac=(len(lst)/len(first_graph))
print("The fraction of the friendship",frac)
        
```

```python
#for the star graph
star_graph=nx.star_graph(n=20)
nx.draw(star_graph,with_labels=True)
mp.show()
```

```python
# the avg degree of the entire graph
avg_deg_eg=np.mean([star_graph.degree(node) for node in star_graph.nodes()])
print("The average degree of entire graph is:",avg_deg_eg)
print(nx.info(star_graph))
```

```python
#The average neighbor degree of each node
def find_avg_deg(node,graph):
    n_neigh=graph.neighbors(node)
    avg_ng_deg=np.mean([graph.degree(n) for n in n_neigh])
    return avg_ng_deg
for node in star_graph.nodes():
    avg_deg_sg=find_avg_deg(node,star_graph)
    print("The average neighbor degree",node,":",avg_deg_sg)
```

```python
#getting the friendship paradox effect on star graph
def view_frnd_paradox(graph):
    slist=[]
    
    count=0
    for node in graph.nodes():
        ideg=graph.degree(node)
        avdeg=find_avg_deg(node,graph)
        if(avdeg>ideg):
            slist.append(1)
            
    sfrac=(len(slist)/len(graph))
    return sfrac
frac=view_frnd_paradox(star_graph)
print("The fraction of friendship paradox effect on star_graph:",frac)

```

```python
#Random graph 
rnd_graph=nx.erdos_renyi_graph(n=200, p=0.2)
nx.draw(rnd_graph,with_labels=True)
mp.show()
```

```python
#Average degree of graph for random graph
ravg_deg=np.mean([rnd_graph.degree(node) for node in rnd_graph.nodes()])
print("The average degree of the graph",ravg_deg)
```

```python
# fraction of nodes in the graph have a larger average neighbor degree than their degree?
rfrac=view_frnd_paradox(rnd_graph)
print("The fraction of friendship paradox effect on random_graph:",rfrac)

```

```python
#Scale_free graph
sf_graph = nx.barabasi_albert_graph(n=200, m=3)
nx.draw(sf_graph,with_labels=True)
mp.show()
```

```python
#Average degree of graph for scale free graph
savg_deg=np.mean([sf_graph.degree(node) for node in sf_graph.nodes()])
print("The average degree of the graph",savg_deg)
```

```python
# fraction of nodes in the graph have a larger average neighbor degree than their degree?
scfrac=view_frnd_paradox(sf_graph)
print("The fraction of friendship paradox effect on scale_free_graph:",scfrac)

```

The friendship paradox did occur for the random graph. 
    1. As we know that the friendship paradox holds for a network when the degree of node is less than the average degree of its neighbours (nodes).
    2. Random graphs are generated by adding nodes one by one and randomly adding edges between the nodes with some probability.
    3. So every node has a fixed probability p for any given possible edge between it and the other node.
    4. We have calculated the fraction of effect of friendship paradox in random graph which comes out to be 56.5% which is a reasonable effect of friendship paradox on random graph.
    5. The random graph model has a drawback, it does not consider high clustering that means when tendency of two people who share a friend can also be friends themselves. This is very different from what we observe in real life social networks.
    Hence effect of friendship paradox is less in random graphs as compared to the scale free graphs.
  
    

The friendship paradox also did occur for the scale free graph whose quantatively effect is more than the random graphs. 
    1. Whenever a node is added in scale free graph a fixed number of edges (m) are also added to that node.
    2. Scale free graphs are those which showcase a preferential attachment behaviour 
    that means people who have more friends are likely to get more friends.
