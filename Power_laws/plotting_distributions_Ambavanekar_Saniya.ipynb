```python
import matplotlib.pyplot as plt
from collections import Counter

with open("terrorism.txt",'r') as r1:

          lines=[int(i) for i in r1]
          

#Example using the head and tail            
#print(lines)
#head and tail
head,tail=lines[:5],lines[-5:]

print(head)
print(tail)


```

```python
#Count for head and tail
a=Counter(head)
b=Counter(tail)
print(a)
print(b)
```

```python
#The Number of nodes we have
Number_of_nodes=len(lines)
print(Number_of_nodes)
```

```python
#For the entire importes list
Nk=Counter(lines)
print(Nk)
```

```python
#Separating the list
keys=list(Nk.keys())
values=list(Nk.values())
print(keys)
print(values)
```

```python
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.scatter(keys,values)
plt.show()

plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.plot(keys,values)
plt.show()
```

```python
#Log -Log Scale 
plt.ylim((0.7,10000))
plt.xlabel("Number of Deaths, x")
plt.ylabel("n(x)")
plt.loglog(keys,values, 'o', markersize=3, markerfacecolor='none')
plt.show()
```

```python
#Log binning
import numpy as np

bins=np.logspace(0.0,4.0,num=40)
print(bins)
```

```python
ax = plt.subplot(1,1,1)

ax.set_xlabel('Number of deaths, x')
ax.set_ylabel('p(x)')
ax.hist(lines, bins=bins, normed=True)
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
#CCDF
y1=values
y2=[]
for c in range(0,len(y1)):
    j=sum(y1[c:])
    y2.append(j)
#y2=y2/y2[0]
#print(values)
newlst=[y2[c]/y2[0] for c in range(0,len(y2))]
print(newlst)
```

```python
plt.loglog(keys,newlst, 'o', markersize=3, markerfacecolor='none')
plt.show()
```
