```python
import networkx as nx

G=nx.barabasi_albert_graph(2500,8,seed=1)
deg=G.degree()
deg_lst=[x[1] for x in deg]
```

```python
import powerlaw
fit=powerlaw.Fit(deg_lst)
print(fit.xmin)
```

```python
#Drawing the ccdf
import matplotlib.pyplot as plt
fig2=fit.plot_ccdf(color='b')
plt.show()
```

```python
#Scaling Exponent
print(fit.alpha)
```

```python
#Visualizing the power law
fig2=fit.plot_ccdf(color="r")
fit.power_law.plot_ccdf(color='b',linestyle='--',ax=fig2)
plt.xlabel('Degrees')
plt.ylabel('P(k)')
plt.show()
```

```python
#Getting real world distribution American Football data
Gr=nx.read_gml('football.gml')
gr_deg=Gr.degree()
gr_deg_lst=[x[1] for x in gr_deg]
print(gr_deg_lst)
```

```python
#Getting the cutoff
fit1=powerlaw.Fit(gr_deg_lst)
print(fit1.xmin)
```

```python
#Scaling exponent
print(fit1.alpha)
```

```python
#Visualizing power law
fig3=fit1.plot_ccdf(color="r")
fit1.power_law.plot_ccdf(color='b',linestyle='--',ax=fig3)
plt.xlabel('Degrees')
plt.ylabel('P(k)')
plt.show()
```
