```python
import networkx as nx

random_gr=nx.erdos_renyi_graph(1200, 0.008, seed=42)
print(nx.info(random_gr))
```

```python
#How to find the length
L=nx.shortest_path_length(random_gr)
L=dict(L)
print(L[2][1])
```

```python
#Function to determine the sorted shortest path lengths
def sorted_shortest_path_lengths(G):
    length=[]
    for i in range(0,len(G)):
        for j in range(i+1,len(G)):
            k=nx.shortest_path_length(G,source=i,target=j)
            length.append(k)
    length.sort()
    return(length)
```

```python
L1=sorted_shortest_path_lengths(random_gr)
print("The sorted list of paths",L1)
```

```python
#histogram of the list of the shortest lengths
%matplotlib inline
import matplotlib.pyplot as plt
df,bins,count=plt.hist(L1,bins=6)
```
