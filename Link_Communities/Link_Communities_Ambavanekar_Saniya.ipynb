```python
import linkcom
```

```python
import networkx as nx

my_graph=nx.erdos_renyi_graph(100,0.1)

print(type(my_graph).__name__)
```

```python
edge_community,S, D, Dlist=linkcom.cluster(my_graph)
print(edge_community)
```

```python
#putting the communities in edge graph
nx.set_edge_attributes(my_graph,edge_community,"linkcom")
nx.write_gexf(my_graph, "linkcom_graph.gexf")
```

```python
#Net Sci graph
graph=nx.read_pajek("netsience.net")
largest_component=max(nx.connected_component_subgraphs(graph), key=len)
print(largest_component.nodes())
```

```python
#Convert to graph


lg=nx.Graph(largest_component)
print(type(lg).__name__)
```

```python
#Link Communities
e2c,S1,D1,Dlist1=linkcom.cluster(lg)
print(e2c)
```

```python
#Save graph
nx.set_edge_attributes(lg,e2c,"netscilinkcom")
nx.write_gexf(lg, "netsci_graph.gexf")
```
