```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.read_gml('dolphins.gml',label='id')

plt.figure()
nx.draw(G,with_lables=True)
plt.show()

#Degree_Centrality
degCent=nx.degree_centrality(G)
nx.set_node_attributes(G,degCent,"degree_Centrality")

#betweenness Centrality
betCent=nx.betweenness_centrality(G,normalized=True,endpoints=False)
nx.set_node_attributes(G,betCent,"betweenness_Centrality")

#Closeness Centrality
closeCent=nx.closeness_centrality(G)
nx.set_node_attributes(G,closeCent,"closeness_Centrality")

#eigenvector Centrality
eigenCent=nx.eigenvector_centrality(G)
nx.set_node_attributes(G,eigenCent,"eigenvector_Centrality")

#katz Centrality
katzCent=nx.katz_centrality(G,0.1,1.0)
nx.set_node_attributes(G,katzCent,"katz_Centrality")
```

```python
#writing the graph
nx.write_gexf(G,'dolphins.gexf')
```
